```python
#make race values more clear

print(df['RACE'].unique())

race_corrections = {
    'WHITE HISPANIC': 'HISPANIC',
    'BLACK HISPANIC': 'HISPANIC',
    'ASIAN / PACIFIC ISLANDER': 'ASIAN',
    'UNKNOWN / REFUSED': 'UNKNOWN',
    'AMER INDIAN / ALASKAN NATIVE': 'NATIVE AMERICAN'
}

df['RACE'] = df['RACE'].replace(race_corrections)
```

```python
# charge 1-4 TYPE clarification (M = misdemenor, F = felony)

type_correction = {
    'M':'Misdemeanor',
    'F':'Felony',
    'A':'Administrative',
    'R':'Regulatory',
    'V':'Vehicle',
    'C': 'Citation',
     np.nan:'Unknown'
}

df['CHARGE 1 TYPE'] = df['CHARGE 1 TYPE'].replace(type_correction)
df['CHARGE 2 TYPE'] = df['CHARGE 2 TYPE'].replace(type_correction)
df['CHARGE 3 TYPE'] = df['CHARGE 3 TYPE'].replace(type_correction)
df['CHARGE 4 TYPE'] = df['CHARGE 4 TYPE'].replace(type_correction)
```

```python
print(df['CHARGE 1 TYPE'].unique())
print(df['CHARGE 2 TYPE'].unique())
print(df['CHARGE 3 TYPE'].unique())
print(df['CHARGE 4 TYPE'].unique())
```

```python
# charge 1-4 CLASS clarification

class_correction = {
    'A':'Class A Misdemeanor',
    'B':'Class B Misdemeanor',
    'C':'Class C Misdemeanor',
    'M':'Misdemeanor (General)',
    'P':'Petty Offense',
    'L':'Local Ordinance',
    '1':'Class 1 Felony',
    '2':'Class 2 Felony',
    '3':'Class 3 Felony',
    '4':'Class 4 Felony',
    'X':'Class X Felony',
    'U':'Unclassified',
    'Z':'Warrant Only',
    np.nan: 'Unknown'
}

df['CHARGE 1 CLASS'] = df['CHARGE 1 CLASS'].replace(class_correction)
df['CHARGE 2 CLASS'] = df['CHARGE 2 CLASS'].replace(class_correction)
df['CHARGE 3 CLASS'] = df['CHARGE 3 CLASS'].replace(class_correction)
df['CHARGE 4 CLASS'] = df['CHARGE 4 CLASS'].replace(class_correction)
```

```python
print(df['CHARGE 1 CLASS'].unique())
print(df['CHARGE 2 CLASS'].unique())
print(df['CHARGE 3 CLASS'].unique())
print(df['CHARGE 4 CLASS'].unique())
```

```python
# charge 1-4 STATUTE. values like 725 ilcs xxxx can be standardized/split (code/title/version) to just 725 etc

def category_conv(code):
    if pd.isna(code):
        return None
    if code.startswith('720'):
        return 'Criminal Law'
    elif code.startswith('625'):
        return 'Traffic' 
    elif code.startswith('725'):
        return 'Procedure' 
    elif code.startswith('430'):
        return 'Special Statutes' 
    elif code.startswith('8') or code.startswith('9') or code.startswith('10'):
        return 'Local Ordinance' 
    else:
        return 'Other' 

df['CHARGE 1 STANDARDIZE'] = df['CHARGE 1 STATUTE'].apply(category_conv)
df['CHARGE 2 STANDARDIZE'] = df['CHARGE 2 STATUTE'].apply(category_conv)
df['CHARGE 3 STANDARDIZE'] = df['CHARGE 3 STATUTE'].apply(category_conv)
df['CHARGE 4 STANDARDIZE'] = df['CHARGE 4 STATUTE'].apply(category_conv)
```

```python
# charge 1-4 DESCRIPTION. normalize casing, group similar types ('domestic battery' variations for ex)


def description_conv(desc):
    if pd.isna(desc):
        return None
    desc = desc.lower()
    if 'domestic battery' in desc:
        return 'Domestic Battery'
    if 'battery' in desc:
        return 'Battery'
    if 'dui' in desc or 'influence' in desc:
        return 'DUI'
    if 'suspended' in desc and 'license' in desc:
        return 'Suspended License'
    if 'theft' in desc:
        return 'Theft'
    if any(word in desc for word in ['possession', 'cannabis', 'narcotic', 'drug']):
        return 'Drug Offense'
    if 'warrant' in desc:
        return 'Warrant'
    if 'assault' in desc:
        return 'Assault'
    if any(word in desc for word in ['traffic', 'speed']):
        return 'Traffic Violation'
    if any(word in desc for word in ['firearm', 'gun', 'weapon']):
        return 'Weapons Offense'
    if 'burglary' in desc:
        return 'Burglary'
    if 'trespass' in desc:
        return 'Trespassing'
    return 'Other'
```

```python
df['CHARGE 1 DESCRIPTION STND'] = df['CHARGE 1 DESCRIPTION'].apply(description_conv)
df['CHARGE 2 DESCRIPTION STND'] = df['CHARGE 2 DESCRIPTION'].apply(description_conv)
df['CHARGE 3 DESCRIPTION STND'] = df['CHARGE 3 DESCRIPTION'].apply(description_conv)
df['CHARGE 4 DESCRIPTION STND'] = df['CHARGE 4 DESCRIPTION'].apply(description_conv)
```

```python
#split arrest date into two columns of DATE and TIME

df['ARREST DATE'] = pd.to_datetime(df['ARREST DATE'], errors='coerce')

df['ARREST DATE ONLY'] = df['ARREST DATE'].dt.date
df['ARREST TIME ONLY'] = df['ARREST DATE'].dt.time
```

```python
df
```

```python
#simplify ARREST DATE ONLY column into seasons to reduce unique values

df['ARREST DATE ONLY'] = pd.to_datetime(df['ARREST DATE ONLY'], errors='coerce')

df['MONTH'] = df['ARREST DATE ONLY'].dt.month

def get_season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    else:
        return 'Fall'
    
df['SEASON'] = df['MONTH'].apply(get_season)
```

```python
#year column
df['YEAR'] = df['ARREST DATE ONLY'].dt.year
```

```python
#simplify ARREST TIME ONLY column to morning, afternoon, evening, etc.
df['ARREST TIME ONLY'] = pd.to_datetime(df['ARREST TIME ONLY'], format='%H:%M:%S', errors='coerce')

df['HOUR'] = df['ARREST TIME ONLY'].dt.hour

def get_time_of_day(hour):
    if 5 <= hour < 12:
        return 'Morning'
    elif 12 <= hour < 17:
        return 'Afternoon'
    elif 17 <= hour < 21:
        return 'Evening'
    else:
        return 'Night'
    
df['TIME OF DAY'] = df['HOUR'].apply(get_time_of_day)
```

```python
df = df.drop(columns=['CB_NO', 'CASE NUMBER', 'ARREST DATE','CHARGE 1 STATUTE', 'CHARGE 2 STATUTE', 'MONTH', 'HOUR'])
```
