```python
#most common types of crimes
#how do crimes vary by race
#how has crime changed over the years
#time of year where crime rate spikes?
#is there a specific time of day where crime rate spikes?
#most common crimes? F vs M


import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv(r"C:\Users\Ryan Wong\Desktop\Arrests Cleaned.csv")

df
```

```python
#most common types of crimes


charges = df['CHARGE 1 DESCRIPTION STND'].value_counts()

plt.figure(figsize=(10, 6))

sns.barplot(x=charges.values, y=charges.index, palette='Blues_r')

plt.title('Most Common Charges')
plt.xlabel('Number of Arrests')
plt.ylabel('Charge')
plt.show()
```

```python
#how do crimes vary by race

crimes = df['RACE'].value_counts()

plt.figure(figsize=(10, 6))

sns.barplot(x=crimes.values, y=crimes.index, palette='Blues_r')
plt.title('Arrests by Race')
plt.xlabel('Race')
plt.ylabel('Number of Arrests')
plt.show()


plt.show()
```

```python
#how has crime changed over the years

years = df['YEAR'].value_counts().nlargest(11) #not showing 2025 since year has not ended yet

plt.figure(figsize=(10, 5))

sns.barplot(x=years.index, y=years.values, palette='Blues_r')
plt.title('Number of Arrests per Year')
plt.xlabel('Year')
plt.ylabel('Number of Arrests')
plt.show()
```

```python
#time of year where crime rate spikes?

time = df['SEASON'].value_counts()

plt.figure(figsize=(10, 10))

sns.barplot(x=time.values, y=time.index, palette='Blues_r')

plt.title('Arrests by Season')
plt.xlabel('Number of Arrests')
plt.ylabel('Season')
plt.show()
```

```python
#is there a specific time of day where crime rate spikes?

time = df['TIME OF DAY'].value_counts()

plt.figure(figsize=(10, 5))

sns.barplot(x=time.values, y=time.index, palette='Blues_r')

plt.title('Most Frequent Arrest Times')
plt.xlabel('Number of Arrests')
plt.ylabel('Time of Day')
plt.show()
```

```python
crime= df['CHARGE 1 TYPE'].value_counts().nlargest(2)

plt.figure(figsize=(10, 5))

sns.barplot(x=crime.values, y=crime.index, palette='Blues_r')

plt.title('Most Common Crime Types')
plt.xlabel('Number of Arrests')
plt.ylabel('Crime Type')
plt.show()
```

```python
#Findings:
#Warrants, drug offenses, and theft are the three most common chrages in Chicago. I excluded 'other' from the top three because it is too vague and could represent many different other types of charges.
#African americans make up the vast majority of arrests in Chicago
#Crime has significantly decreased since 2014. The graph shows the past 5 years as having significanlty less crime compared to 2014-19
#Across the four seasons, crime rates are fairly consistent, with the highest number of offenses occurring during the summer.
#Most crimes happen as the day moves forward with nighttime having the highest probability of crime to occur. 
#Misdemeanors are around twice as common compared to felony charges.
```
